```python
def dfactorial(n):
    fact = 1
    if n%2 == 0:
        for i in range(2,n+2,2):
            fact *= i
    else:
        for i in range(1,n+2,2):
            fact *= i
    return fact

dfactorial(5)
```

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

def sf(n):
    p = 1
    for i in range(1,n+1):
        p *= factorial(i)
    return p

sf(4)
```

Output: index of max number in the list

```python
def maxId(L):
    l = list(map(int, L))
    n = 0
    for i,j in enumerate(l):
        if n < j:
            p = i
            n = j
    return p

maxId([1, 2, '42', '3', '4', '5', 6, 13])
```

### 3.2.10

```python
def is_prime(n):
    d = 0
    if n < 10:
        for i in range(1,n+1):
            if n%i == 0:
                d +=1
        return d == 2
    else:
        for i in range(1,11):
            if n%i == 0:
                d +=1
        return d == 1
```

### 8.2.7

Ribon with length L was cutted into N pieces. What is prbobility that 1 piece is equal or longer then M

```python
import random
import numpy as np

L, N, M = [int(input()) for i in range(3)]

results = []

for i in range(100000):
    cuts = np.sort(np.random.uniform(low = 0, high = L, size = N-1))
    length =[cuts[0]]
    for j in range(1, N-1):
        length.append(cuts[j]-cuts[j-1])
    length.append(L-cuts[-1])
    if max(length) >= M:
        results.append(1)
    else:
        results.append(0)

print(f'{round(sum(results)/len(results)*100,1)}%')
```
