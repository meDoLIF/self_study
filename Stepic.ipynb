Если оба числа равны 0, выведите на печать строку "Empty Ints"
Иначе выведите на печать сумму этих чисел


Если получен тип 'str':

Считайте ещё 1 строку.

Если строка пустая, выведите на печать строку "Empty String"
Иначе напечатайте эту строку


Если получен тип 'list':

Считайте 1 строку и разрежьте её на элементы с помощью split().

Если получившийся список пустой, выведите на печать строку "Empty List"
Иначе выведите на печать последний элемент списка. Примечание, используйте отрицательный индекс.

Используйте приведение типов для определения пустых списков и строк, а так же чисел, равных 0.

```python
a = input()

if a == 'int':
    b = int(input())
    c = int(input())
    if b and c:
        print(b+c) 
    else:
        print('Empty Ints') 
elif a =='str':
    b = input()
    if b:
        print(b)
    else:
        print('Empty String')
elif a =='list':
    b = input().split()
    if b:
        print(b[-1])
    else:
        print('Empty List')
else:
    print('Unknown type')
```

```python
def dfactorial(n):
    fact = 1
    if n%2 == 0:
        for i in range(2,n+2,2):
            fact *= i
    else:
        for i in range(1,n+2,2):
            fact *= i
    return fact

dfactorial(5)
```

### 2.7.7

Make function for factorial and superfactorial

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

def sf(n):
    p = 1
    for i in range(1,n+1):
        p *= factorial(i)
    return p

sf(4)
```

### 2.7.8

Create function
input: list with numbers (type can be str)\
Output: index of max number in the list

```python
def maxId(L):
    l = list(map(int, L))
    n = 0
    for i,j in enumerate(l):
        if n < j:
            p = i
            n = j
    return p

maxId([1, 2, '42', '3', '4', '5', 6, 13])
```

### 3.2.9

```python
def fib(n):
    l = [0,1]
    i = 1
    while i < n:
        l.append(l[i]+l[i-1])
        i += 1
    return l[n]
```

### 3.2.10

Check if number is prime

```python
def is_prime(n):
    d = 0
    if n < 10:
        for i in range(1,n+1):
            if n%i == 0:
                d +=1
        return d == 2
    else:
        for i in range(1,11):
            if n%i == 0:
                d +=1
        return d == 1
```

Luka's sequence

```python
def luka(L0, L1, n):
    l = [L0, L1]
    if n < 2:
        return l[n]
    else:
        for i in range(2,n):
            L0,L1 = L1, L0+L1
        return L0+L1
```
