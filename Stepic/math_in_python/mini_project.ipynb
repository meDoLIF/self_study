```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x, a, b):
    return (x**b + a**b) / x**b

def f_zero(x):
    return x*0
```

```python
neg_x = np.arange(-4, -0.02, 0.02)
pos_x = np.arange(0.02, 4, 0.02)
```

# Задача 1 #

```python
fig, axes = plt.subplots(figsize=(11, 7))
axes.plot(neg_x, f(neg_x, a=1, b=1), 'r', pos_x, f(pos_x, a=1, b=1), 'r')
axes.plot(neg_x, f(neg_x, a=2, b=1), 'b', pos_x, f(pos_x, a=2, b=1), 'b')
axes.plot(neg_x, f(neg_x, a=1, b=2), 'g', pos_x, f(pos_x, a=1, b=2), 'g')

axes.set_xlabel('x', fontsize=13)
axes.set_ylabel('f(x)', fontsize=13)
axes.set_title('f(x) = $\\frac{x^b + a^b}{x^b}$', fontsize=16)
axes.legend(['a=1, b=1', '_hidden', 'a=2, b=1', '_hidden', 'a=1, b=2', '_hidden'])

axes.set_ylim(-40, 40)
```

```python
pos_x_zero = np.arange(1.5, 0.001, -0.001)
pos_x_inf = np.arange(4, 30, 0.1)
```

```python
fig, axes = plt.subplots(figsize=(12, 8))
axes.plot(pos_x, f(pos_x, a=1, b=1), 'r')
axes.plot(pos_x, f(pos_x, a=2, b=1), 'b')
axes.plot(pos_x, f(pos_x, a=1, b=2), 'g')

axes.set_xlabel('x', fontsize=13)
axes.set_ylabel('f(x)', fontsize=13)
axes.set_title('f(x) = $\\frac{x^b + a^b}{x^b}$', fontsize=16)
axes.legend(['a=1, b=1', 'a=2, b=1', 'a=1, b=2'], loc=3)

axes.set_ylim(-10, 40)

axes_2 = fig.add_axes([0.27, 0.55, 0.25, 0.25])
axes_3 = fig.add_axes([0.6, 0.55, 0.25, 0.25])

axes_2.plot(pos_x_zero, f(pos_x_zero, a=1, b=1), 'r')
axes_2.plot(pos_x_zero, f(pos_x_zero, a=2, b=1), 'b')
axes_2.plot(pos_x_zero, f(pos_x_zero, a=1, b=2), 'g')
axes_2.set_ylim(-1, 7)

axes_3.plot(pos_x_inf, f(pos_x_inf, a=1, b=1), 'r')
axes_3.plot(pos_x_inf, f(pos_x_inf, a=2, b=1), 'b')
axes_3.plot(pos_x_inf, f(pos_x_inf, a=1, b=2), 'g')
axes_3.set_ylim(0.75, 1.75)
```

```python
neg_x_inf = np.arange(-90, -120, -0.1)
zero = np.array([-150, 20])
```

```python
fig, axes = plt.subplots(figsize=(12, 8))
axes.plot(neg_x, f(neg_x, a=1, b=1), 'r')
axes.plot(neg_x, f(neg_x, a=2, b=1), 'b')
axes.plot(neg_x, f(neg_x, a=1, b=2), 'g')
axes.plot(zero, f_zero(zero), 'black')

axes.set_xlabel('x', fontsize=13)
axes.set_ylabel('f(x)', fontsize=13)
axes.set_title('f(x) = $\\frac{x^b + a^b}{x^b}$', fontsize=16)
axes.legend(['a=1, b=1', 'a=2, b=1', 'a=1, b=2'], loc=3)
axes.set_ylim(-10, 30)
axes.set_xlim(-3, 0)

axes_2 = fig.add_axes([0.3, 0.45, 0.35, 0.35])

axes_2.plot(neg_x_inf, f(neg_x_inf, a=1, b=1), 'r')
axes_2.plot(neg_x_inf, f(neg_x_inf, a=2, b=1), 'b')
axes_2.plot(neg_x_inf, f(neg_x_inf, a=1, b=2), 'g')
axes_2.set_ylim(0.95, 1.05)
```

# Задача 2 #

```python
x = np.arange(-2, 3, 0.02)
```

```python
fig, axes = plt.subplots(figsize=(11, 7))

axes.set_xlabel('x', fontsize=13)
axes.set_ylabel('f(x)', fontsize=13)
axes.set_title('f(x) = 1+ $(\\frac{a}{x})^b$', fontsize=16)
axes.set_ylim(-2, 5)

axes.plot(x, f(x, a=1, b=0), 'r', alpha=0.6)
axes.plot(x, f(x, a=0, b=0), '--b', alpha=0.7)
axes.plot(x, f(x, a=0, b=1), '--g', alpha=0.7)
axes.plot(x, f(x, a=0, b=2), 'c', alpha=0.6)
axes.plot(x, f(x, a=1, b=-1), 'm')
axes.plot(x, f(x, a=2, b=-1), 'y')

axes.legend(['a=1, b=0', 'a=0, b=0', 'a=0, b=1', 'a=0, b=2', 'a=1, b=-1', 'a=2, b=-1'])
```

Выводы:

1. При a=1, b=0: f(x)=2
2. При a=0, b=0: f(x)=1
3. Функция при a=1, b=-1 возврастает быстрее, чем функция при a=2, b=-1
4. Наклонные функции пересекаются в точке (0, 1)
5. Функция f(x) при a=1, b=-1 пересекается с функциями f(x)=2 в точке (1, 2)
6. Функция f(x) при a=2, b=-1 пересекается с функциями f(x)=2 в точке (2, 2)

# Задача 4 #

```python
def f(x, b, a=1):
    return (x**b + a**b) / x**b
```

```python
pos_x = np.arange(0.001, 2, 0.001)
```

```python
fig = plt.figure(figsize=(12, 8))


ax1 = fig.add_axes([0, 1.2, 0.4, 0.4])
ax1.set_xlabel('x', fontsize=13)
ax1.set_ylabel('f(x)', fontsize=13)
ax1.set_title('f(x) = 1+ $(\\frac{1}{x})^b$', fontsize=16)
ax1.set_ylim(-1, 6)

ax1.plot(pos_x, f(pos_x, b=0), 'b--')
ax1.plot(pos_x, f(pos_x, b=-1), 'r--')
ax1.plot(pos_x, f(pos_x, b=0.5), 'g')
ax1.plot(pos_x, f(pos_x, b=0.8), 'm')

ax1.legend(['b=0', 'b=-1', 'b=0.5', 'b=0.8'], fontsize=7)
#--------------------------------------------------------------

ax2 = fig.add_axes([0, 0.6, 0.4, 0.4])
ax2.set_xlabel('x', fontsize=13)
ax2.set_ylabel('f(x)', fontsize=13)
ax2.set_title('f(x) = 1+ $(\\frac{1}{x})^b$', fontsize=16)
ax2.set_ylim(-1, 6)

ax2.plot(pos_x, f(pos_x, b=0), 'b--')
ax2.plot(pos_x, f(pos_x, b=-1), 'r--')
ax2.plot(pos_x, f(pos_x, b=-0.5), 'g')
ax2.plot(pos_x, f(pos_x, b=-0.8), 'm')

ax2.legend(['b=0', 'b=-1', 'b=-0.5', 'b=-0.8'], fontsize=7)
#--------------------------------------------------------------

ax3 = fig.add_axes([0, 0, 0.4, 0.4])
ax3.set_xlabel('x', fontsize=13)
ax3.set_ylabel('f(x)', fontsize=13)
ax3.set_title('f(x) = 1+ $(\\frac{1}{x})^b$', fontsize=16)
ax3.set_ylim(-1, 6)

ax3.plot(pos_x, f(pos_x, b=0), 'b--')
ax3.plot(pos_x, f(pos_x, b=-1), 'r--')
ax3.plot(pos_x, f(pos_x, b=-1.5), 'g')
ax3.plot(pos_x, f(pos_x, b=-2.5), 'm')

ax3.legend(['b=0', 'b=-1', 'b=-1.5', 'b=-2.5'], fontsize=7)
```

Вывод:
Построение данных графиков показывает, что при любом значении b и a=1, функции пересекаются в точке (1, 2)
