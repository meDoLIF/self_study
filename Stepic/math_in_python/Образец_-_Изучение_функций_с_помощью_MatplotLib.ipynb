## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, a, b):
    return (x**b + a**b)/x**b

%matplotlib inline

x = np.arange(-30, 31)

fig, axes = plt.subplots(figsize = (25,18))
axes.plot(x, f(x, 1, 1), 'r', label = '$\\alpha = 1, \\beta = 1$')
axes.plot(x, f(x, 2, 1), 'b', label = '$\\alpha = 2, \\beta = 1$')
axes.plot(x, f(x, 1, 2), 'g', label = '$\\alpha = 1, \\beta = 2$')

axes.set_title('$f(x) = \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta} }$ for diffrent $\\alpha$ и $\\beta$.', fontsize = 25)

axes.set_xlabel('x', fontsize = 18)
axes.set_ylabel('f(x)', fontsize = 18)
axes.legend(fontsize = 18)
```

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
fig = plt.figure(figsize=(13,8), dpi=120)

axes1 = fig.add_axes([0, 0, 1, 1])
x = np.arange(-30, 31)
axes1.plot(x, f(x, 1, 1), 'r', label = '$\\alpha = 1, \\beta = 1$')
axes1.plot(x, f(x, 2, 1), 'b', label = '$\\alpha = 2, \\beta = 1$')
axes1.plot(x, f(x, 1, 2), 'g', label = '$\\alpha = 1, \\beta = 2$')

axes1.set_title('$f(x) = \\frac{x^{\\beta} + \\alpha^{\\beta}}{x^{\\beta} }$ for diffrent $\\alpha$ и $\\beta$.', fontsize = 25)

axes1.set_xlabel('x', fontsize = 18)
axes1.set_ylabel('f(x)', fontsize = 18)
axes1.legend(fontsize = 18)

axes2 = fig.add_axes([0.05, 0.65, 0.3, 0.3])

x = np.arange(0, 1, 0.01)
axes2.plot(x, f(x, 1, 1), 'r')
axes2.plot(x, f(x, 2, 1), 'b')
axes2.plot(x, f(x, 1, 2), 'g')

axes2.set_title('For small X', fontsize = 10)

axes3 = fig.add_axes([0.05, 0.05, 0.3, 0.3])

x = np.arange(100, 1000, 10)
axes3.plot(x, f(x, 1, 1), 'r')
axes3.plot(x, f(x, 2, 1), 'b')
axes3.plot(x, f(x, 1, 2), 'g')

axes3.set_title('For big X', fontsize = 10)
```
