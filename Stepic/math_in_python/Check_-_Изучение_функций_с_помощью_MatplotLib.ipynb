## –ó–∞–¥–∞—á–∞ ‚Ññ1

### –ß–∞—Å—Ç—å 1

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≤ –æ–±—â–∏—Ö –æ—Å—è—Ö –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
x = np.linspace(-1000,1000,1000001)
fx_name = r'$f(x)=\frac{x^ùõΩ + ùõº^ùõΩ}{x^ùõΩ}$'
def func(x, ùõº, ùõΩ):
    with np.errstate(divide='ignore', invalid='ignore'):
        return (x ** ùõΩ + ùõº ** ùõΩ) / x ** ùõΩ


y1 = func(x, 1, 1)
y2 = func(x, 2, 1)
y3 = func(x, 1, 2)
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, y1, label='$ùõº=1, ùõΩ=1$')
axes.plot(x, y2, label='$ùõº=2, ùõΩ=1$')
axes.plot(x, y3, label='$ùõº=1, ùõΩ=2$')
axes.legend()
axes.set_ybound(lower=-80, upper=80)
axes.set_xbound(lower=-6, upper=6)
axes.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ {fx_name} –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö $ùõº$ –∏ $ùõΩ$')
axes.set_xlabel('x', fontsize=16)
axes.set_ylabel('f(x)', fontsize=16)
plt.savefig('1_0.svg')
```

### –ß–∞—Å—Ç—å 3

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≤ –æ–±—â–∏—Ö –æ—Å—è—Ö –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è $x<0$.

–ù–∞ —Ç–æ–º –∂–µ –≥—Ä–∞—Ñ–∏–∫–µ —Å–¥–µ–ª–∞—Ç—å 1 –≤—Ä–µ–∑–∫—É, –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—â—É—é –ø–æ–≤–µ–¥–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ x –æ—Ç 0 –∫ $-\infty$.

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä–æ–≤–∞—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å (–∏–ª–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å) –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π –∏ —Å—Ç—Ä–µ–º–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π. –¢–∞–∫ –∂–µ –Ω–∞–Ω–µ—Å–∏—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∏ –ø—Ä—è–º—É—é f(x) = 0.

–¶–≤–µ—Ç –ª–∏–Ω–∏–π –Ω–∞ –≤—Ä–µ–∑–∫–∞—Ö –∏ –æ—Å–Ω–æ–≤–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º –¥–ª—è –æ–¥–Ω–∏—Ö –∏ —Ç–µ—Ö –∂–µ –∑–Ω–∞—á–µ–Ω–∏–π $\alpha$ –∏ $\beta$.

```python
x = np.linspace(-1000,0,100001)
x2 = np.linspace(-1000,1,100001)
fx_name = r'$f(x)=\frac{x^ùõΩ + ùõº^ùõΩ}{x^ùõΩ}$'
def func(x, ùõº, ùõΩ):
    with np.errstate(divide='ignore', invalid='ignore'):
        return (x ** ùõΩ + ùõº ** ùõΩ) / x ** ùõΩ
def f0(x):
    return x*0

y1 = func(x, 1, 1)
y2 = func(x, 2, 1)
y3 = func(x, 1, 2)
y4 = f0(x2)
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, y1, label='$ùõº=1, ùõΩ=1$')
axes.plot(x, y2, label='$ùõº=2, ùõΩ=1$')
axes.plot(x, y3, label='$ùõº=1, ùõΩ=2$')
axes.plot(x2, y4, label='f(x)=0')
axes.legend()
axes.set_ybound(lower=-80, upper=80)
axes.set_xbound(lower=-6, upper=1)
axes.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ {fx_name} –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö $ùõº$ –∏ $ùõΩ$')
axes.set_xlabel('x', fontsize=16)
axes.set_ylabel('f(x)', fontsize=16)
axes2 = fig.add_axes([0.2, 0.6, 0.25, 0.25])
axes2.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏ x -> -$‚àû$', fontsize=8)
axes2.plot(x, y1, label='$ùõº=1, ùõΩ=1$')
axes2.plot(x, y2, label='$ùõº=2, ùõΩ=1$')
axes2.plot(x, y3, label='$ùõº=1, ùõΩ=2$')
axes2.plot(x2, y4, label='f(x)=0')
axes2.set_ybound(lower=0.95, upper=1.01)
axes2.set_xbound(lower=-1000, upper=-99)
axes2.legend()
axes3 = fig.add_axes([0.2, 0.2, 0.25, 0.25])
axes3.plot(x, y1, label='$ùõº=1, ùõΩ=1$')
axes3.plot(x, y2, label='$ùõº=2, ùõΩ=1$')
axes3.plot(x, y3, label='$ùõº=1, ùõΩ=2$')
axes3.plot(x2, y4, label='f(x)=0')
axes3.legend()

axes3.set_ybound(lower=-5, upper=3)
axes3.set_xbound(lower=-3, upper=0)
axes3.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏ –º–∞–ª—ã—Ö x', fontsize=8)
plt.savefig('1_2.svg')
```

ùõº=1, ùõΩ=2 –º–µ–Ω—è–µ—Ç –ø–æ–≤–µ–¥–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏ x -> 0 —Å–ª–µ–≤–∞ –æ—Ç –æ—Å–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç.
ùõº=1, ùõΩ=2 –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç –æ—Å—å –∞–±—Å—Ü–∏—Å—Å –ø—Ä–∏ —Ö= -1.0,

–ü—Ä–∏ —Ö-> ‚àí‚àû –∏ +‚àû –≥—Ä–∞—Ñ–∏–∫–∏ —Å—Ç—Ä–µ–º—è—Ç—Å—è –∫ 1.

## –ó–∞–¥–∞—á–∞ ‚Ññ2

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≤ –æ–±—â–∏—Ö –æ—Å—è—Ö –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

–°–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã –æ –ø–æ–≤–µ–¥–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤

```python
x = np.linspace(-100,100,10001)
x2 = np.linspace(-100,99,10001)
fx_name = r'$f(x)=\frac{x^ùõΩ + ùõº^ùõΩ}{x^ùõΩ}$'
def func(x, ùõº, ùõΩ):
    with np.errstate(divide='ignore', invalid='ignore'):
        return (x ** ùõΩ + ùõº ** ùõΩ) / x ** ùõΩ


y1 = func(x, 1, 0)
y2 = func(x, 0, 0)
y3 = func(x, 0, 1)
y4 = func(x, 0, 2)
y5 = func(x, 1, -1)
y6 = func(x, 2, -1)
```

```python
fig, axes = plt.subplots(figsize=(13, 8))
axes.plot(x, y1, label='$ùõº=1, ùõΩ=0$')
axes.plot(x2, y2, label='$ùõº=0, ùõΩ=0$')
axes.plot(x, y3, label='$ùõº=0, ùõΩ=1$')
axes.plot(x2, y4, label='$ùõº=0, ùõΩ=2$')
axes.plot(x, y5, label='$ùõº=1, ùõΩ=-1$')
axes.plot(x, y6, label='$ùõº=2, ùõΩ=-1$')
axes.legend()
axes.set_ybound(lower=-80, upper=80)
axes.set_xbound(lower=-100, upper=100)
axes.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ {fx_name} –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö $ùõº$ –∏ $ùõΩ$')
axes.set_xlabel('x', fontsize=16)
axes.set_ylabel('f(x)', fontsize=16)
axes2 = fig.add_axes([0.3, 0.55, 0.2, 0.3])
axes2.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏ –±–æ–ª—å—à–∏—Ö x', fontsize=8)
axes2.plot(x, y1, label='$ùõº=1, ùõΩ=0$')
axes2.plot(x2, y2, label='$ùõº=0, ùõΩ=0$')
axes2.plot(x, y3, label='$ùõº=0, ùõΩ=1$')
axes2.plot(x2, y4, label='$ùõº=0, ùõΩ=2$')
axes2.plot(x, y5, label='$ùõº=1, ùõΩ=-1$')
axes2.plot(x, y6, label='$ùõº=2, ùõΩ=-1$')
axes2.legend(fontsize=6)
axes2.set_ybound(lower=-0.9, upper=2.5)
axes2.set_xbound(lower=98, upper=100)
axes3 = fig.add_axes([0.55, 0.15, 0.3, 0.3])
axes3.plot(x, y1, label='$ùõº=1, ùõΩ=0$')
axes3.plot(x2, y2, label='$ùõº=0, ùõΩ=0$')
axes3.plot(x, y3, label='$ùõº=0, ùõΩ=1$')
axes3.plot(x2, y4, label='$ùõº=0, ùõΩ=2$')
axes3.plot(x, y5, label='$ùõº=1, ùõΩ=-1$')
axes3.plot(x, y6, label='$ùõº=2, ùõΩ=-1$')
axes3.legend(fontsize=6)
axes3.set_ybound(lower=-2, upper=6)
axes3.set_xbound(lower=-2, upper=5)
axes3.set_title(f'–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏ –º–∞–ª—ã—Ö x', fontsize=8)
plt.savefig('2_0.svg')
```

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$
